## `THE DATASET`: [Sustainable Development Data 2000-2022](https://www.kaggle.com/datasets/sazidthe1/sustainable-development-report)
- The file to use: 'sdg_index_2000-2022.csv'

```python
import pandas as pd

#load the data:
df = pd.read_csv('sdg_index_2000-2022.csv')
df.head()
```

```python
#examine the data:
df.info()
```

> ### Cast the variables country, year, countrycode into string:

```python
df[['country', 'year', 'country_code']] = df[['country', 'year', 'country_code']].astype('string')
df[['country', 'year', 'country_code']].info()
```

> ### Extract the data of year 2022

```python
df22 = df[df['year'] == '2022']
df22.reset_index(inplace = True)
df22
```

```python
#drop the column index:
df22.drop(columns = ['index'], inplace = True)
df22
```

```python
#find out  the correlation between the total sdg_index_score and the index score of each goal:
score = df22[['sdg_index_score', 'goal_1_score',
       'goal_2_score', 'goal_3_score', 'goal_4_score', 'goal_5_score',
       'goal_6_score', 'goal_7_score', 'goal_8_score', 'goal_9_score',
       'goal_10_score', 'goal_11_score', 'goal_12_score', 'goal_13_score',
       'goal_14_score', 'goal_15_score', 'goal_16_score', 'goal_17_score']]
score
```

### 1. Find the correlation between any two variables:

```python
corr = score.corr()
corr
```

```python
for x in corr.columns:
    vrange = corr[x][corr[x].index != x] #exclude the std between the var and itself
    highest = vrange.max() #find out the max value
    Y = vrange[vrange == highest].index.tolist()[0] #to extract the name of the index from the pd.series object
    print(f'The variable [{x}] has the higest correlation with the variable [{Y}] out of all others, with the correlation coefficient of {highest}.')
    print()
```

### 2. Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
#the heatmap: 
sns.heatmap(data = corr, #the data
            annot=False,  #annotate the data or not
            cmap='coolwarm')  #the color map

#label the graph:
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('The heatmap of the sdg index scores')

#size:
plt.figure(figsize = (30,20))
plt.show()
```

### 3. Boxplot:

```python
#the boxplot:
sns.boxplot(score)

#label the graph:
plt.xlabel('Variables')
plt.ylabel('value Range')
plt.title('The boxplots of the sdg index scores')

#size:
plt.figure(figsize = (30,20)) #length, width
plt.show()
```

```python
#the boxplot:
sns.violinplot(score)

#label the graph:
plt.xlabel('Variables')
plt.ylabel('value Range')
plt.title('The violin plots of the sdg index scores')

#size:
plt.figure(figsize = (30,20)) #length, width
plt.show()
```
