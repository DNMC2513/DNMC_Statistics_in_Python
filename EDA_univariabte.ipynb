> ### EDA of the numerical variables:

```python
num = df22.describe()
num
```

### 1) Mean_ Measure/Estimate of location:

```python
#Mean:
for x in num.columns:
    a = num.loc['mean', x]
    print(f'The mean value of {x} is: {a}')
    print()
```

### 2)Median_ Measure/Estimate of location:

```python
for x in num.columns:
    a = num.loc['50%',x]
    print(f'The median value of {x} is: {a}')
    print()
```

### 3) Interquartile range _  Measure/Estimate of dispersion:

```python
for x in num.columns:
    f = num.loc['25%', x]
    t = num.loc['75%', x]
    iqr = t-f
    ma = num.loc['max',x]
    mi = num.loc['min',x]
    
    print(f'The interquartile range of {x} in 2022:')
    print(f'The max value of the range is: {ma}')
    print(f'The min value of the range is: {mi}')
    print(f'- The first quartile is: {f}')
    print(f'- The third quartile is: {t}')
    print(f'- The interquartile range is: {iqr}')
    print(f'- Outliers are values that falls below {f - 1.5*iqr} and above {t + 1.5*iqr} in the range')
    print()
```

### 4.Trimmed mean

```python
#sort values:
sorted_score = df22['sdg_index_score'].sort_values().reset_index()
sorted_score
```

```python
#get the numpy array of the variable of interest:
sorted_score = sorted_score['sdg_index_score']
sorted_score
```

```python
#trim off the lowest 5% and highest 5% of the data range
number_of_trimmedvalues = int(len(sorted_score) * 0.05)
number_of_trimmedvalues
```

```python
#trim the data range:
trimmed_score = sorted_score[number_of_trimmedvalues : - number_of_trimmedvalues]
#the trimmed range starts at the 9-indexed element 
#stop at the value before the -9-indexed number
trimmed_score
```

```python
trimmed_mean = trimmed_score.mean()
trimmed_mean
```
